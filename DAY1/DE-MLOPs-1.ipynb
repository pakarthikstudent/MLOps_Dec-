```python
URL = 'https://webscraper.io/test-sites/e-commerce/static/computers/laptops'
```

```python
response = requests.get(URL)
soup_obj = bs4.BeautifulSoup(response.text)
```

```python
products = soup_obj.find_all("div",class_="thumbnail")
```

```python
import csv
```

```python
with open("products.csv","w", newline="",encoding="utf-8") as wobj:
    writer = csv.writer(wobj)
    writer.writerow(["Product Name","Price","Description"])

    for item in products:
        name = item.find("a",class_="title")
        price = item.find("a",class_="pull-right-price")
        description = item.find("a",class_="description")

        print(name,price)
        writer.writerow([name,price,description])

    
```
